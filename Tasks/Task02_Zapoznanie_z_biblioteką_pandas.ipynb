#### 🌐 Pobieranie danych z internetu za pomocą `requests` i `StringIO` w Pythonie

#### 📥 Biblioteka `requests` – pobieranie zasobów

Biblioteka **`requests`** w Pythonie jest bardzo przydatna do pobierania danych z internetu. Pozwala wysyłać zapytania HTTP, aby uzyskać dostęp do plików, obrazów, tekstów, czy zawartości stron internetowych.

#### ✨ Jak działa `requests.get(url)`?

1. **Wysyła zapytanie HTTP**: `requests.get(url)` wysyła zapytanie typu **GET** do serwera, prosząc o dostęp do określonego zasobu.
2. **Zwraca odpowiedź**: Serwer zwraca zasób, który przechowujemy w obiekcie `response`.

Przykład pobrania pliku:
```python
import requests

url = 'https://example.com/file.csv'
response = requests.get(url)
```

####📄 StringIO – przekształcanie tekstu w "wirtualny plik"
Załóżmy, że mamy dane jako tekst i chcemy je przekształcić na tabelę danych, ale bez tworzenia fizycznego pliku.

🔍 Co robi StringIO?

-Przechowuje dane w pamięci RAM.
-Pozwala innym bibliotekom (np. pandas) traktować dane tekstowe jak plik, co umożliwia ich szybkie wczytanie.

````python
import pandas as pd
import requests
from io import StringIO

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import boxcox, skew

url = 'https://drive.google.com/uc?export=download&id=1Myi9ZzUu7qE5tU4KW102rSLTd6xGGAN4'


response = requests.get(url)
if response.status_code == 200:
    data = StringIO(response.text)
    data_heart = pd.read_csv(data)
    print(data_heart.head())
    print(data_heart.describe())

````

**Jak działa mapowanie?** 🛠️
